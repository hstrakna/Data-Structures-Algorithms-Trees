```python
class BST(object):
    def __init__(self, val):
        self.left = None
        self.right = None
        self.totDepths = 0
        self.depth = 0
        self.nodeCount = 0
        self.val = val

    def insertNode(self, newNode):
        
        if self.val:
            if newNode < self.val:
                if self.left == None:
                    #create new node
                    self.left = BST(newNode)
                    self.depth+=1
                else:
                    #travel along left
                    self.left.insertNode(newNode)
                    self.depth+=1
            elif newNode > self.val:
                if self.right == None:
                    self.right = BST(newNode)
                    self.depth+=1
                else:
                    self.right.insertNode(newNode)
                    self.depth+=1
            #adjust node count        
            self.nodeCount += 1
            
            #adjust total depth of node combined
            self.totDepths += self.depth
        else:
            self.val = newNode
    
    #helper functions
    def totalDepths(self):
        return self.totDepths

    def nodeCount(self):
        return self.nodeCount
    
    def avgDepth(self):
        return self.totDepths//self.nodeCount
    
```

```python
import random

array = []
for i in range(0, 5000):
    array.append(random.randrange(0,50000))

root = array[0]
tree = BST(root)
for val in array:
    tree.insertNode(val)


print("Average Depth of Nodes: {}".format(tree.avgDepth()))
```

```python
class Node(object):
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None
        self.depth = 0
        self.depthTotal = 0
        self.nodeCount = 0
        self.SRCount = 0
        self.DRCount = 0

class AVLTree(object):
    
    #rotation count
    SRCount = 0
    DRCount = 0
    def insertNode(self, root, newNode):
        
        #create new root if there is none
        if not root:
            return Node(newNode)
        
        #insert on left or right
        elif newNode < root.val:
            root.left = self.insertNode(root.left, newNode)
        elif newNode > root.val:
            root.right = self.insertNode(root.right, newNode)

        # set the root's depth to the maximum of its children plus one

        root.depth = 1 + max(self.getDepth(root.left), self.getDepth(root.right))

        #cases for zig and zag

        if self.getDepth(root) > self.getDepth(root.right) + 1 and newNode < root.left.val:
            self.SRCount += 1

            return self.zag(root)

        if self.getDepth(root) > self.getDepth(root.right) + 1 and newNode < root.left.val:
            self.DRCount += 1

            root.left = self.zig(root.left)
            return self.zag(root)

        if self.getDepth(root) > self.getDepth(root.left) + 1 and newNode > root.right.val:
            self.SRCount += 1
            return self.zig(root)

        if self.getDepth(root) > self.getDepth(root.left) + 1 and newNode < root.right.val:
            self.DRCount += 1
            root.right = self.zag(root.right)
            return self.zig(root)
        #adjust depth and node count totals
        root.depthTotal += root.depth
        root.nodeCount += 1
        print(root.depth)
        return root


    #zig and zag functions
    def zig(self, b):
        # set node a to b's right child
        a = b.right
        # get a's left child and set it to childLeft
        childLeft = a.left

        # swtich a's left child to b
        a.left = b
        # switch b's right child to a's left child
        b.right = childLeft

        # set b's height to the maximum of its children plus one
        b.depth = max(self.getDepth(b.left), self.getDepth(b.right)) + 1
        a.depth = max(self.getDepth(a.left), self.getDepth(a.right)) + 1
        # return new root
        return a

    # do the inverse of the zig operation
    def zag(self, b):
        a = b.left
        childRight = a.right

        a.right = b
        b.left = childRight

        b.depth = max(self.getDepth(b.left), self.getDepth(b.right))
        a.depth = max(self.getDepth(a.left), self.getDepth(a.right))

        return a

    #helper functions
    def getDepth(self, root):
        if not root:
            return 0
        return root.depth

    def avgDepth(self):
        if not root:
            return 0
        avg = root.depthTotal // root.nodeCount
        return avg

    def singleRotationCount(self):
        return self.SRCount

    def doubleRotationCount(self):
        return self.DRCount


```
